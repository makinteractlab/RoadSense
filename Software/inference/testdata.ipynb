```python
import os
import numpy as np
import librosa
import librosa.display
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
asdir = 'C:/Users/MAKinteract/makinteract/as/'
cedir = 'C:/Users/MAKinteract/makinteract/ce/'
assavedir = 'C:/Users/MAKinteract/makinteract/asplot'
cesavedir = 'C:/Users/MAKinteract/makinteract/ceplot'
if not os.path.exists(assavedir):
    os.mkdir(assavedir)
if not os.path.exists(cesavedir):
    os.mkdir(cesavedir)

aslist = [os.path.join(asdir,f) for f in os.listdir(asdir) if os.path.isfile(os.path.join(asdir,f))]
asnamelist = [f for f in os.listdir(asdir) if os.path.isfile(os.path.join(asdir,f))]
celist = [os.path.join(cedir, f) for f in os.listdir(cedir) if os.path.isfile(os.path.join(cedir, f))]
cenamelist = [f for f in os.listdir(cedir) if os.path.isfile(os.path.join(cedir, f))]
```

```python
aslist
aslen = len(aslist)
print(aslen)
```

```python
celist
celen = len(celist)
print(celen)
```

```python
tempce = []
for i in range(celen):
    if int(cenamelist[i].split(".")[0].split("_")[1]) >= 71:
        tempce.append(celist[i])

tempas = []
for i in range(aslen):
    if int(asnamelist[i].split(".")[0].split("_")[1]) >= 54:
        tempas.append(aslist[i])
```

```python
for i in tqdm(range(len(tempce))):
    ns = np.load(tempce[i])
    S_DB = librosa.power_to_db(ns, ref=np.max)
    librosa.display.specshow(S_DB,sr=10000, hop_length=512, 
                             x_axis='time', y_axis='mel');
    savename = tempce[i].split("/")[-1].split(".")[0] + ".png"
    savedir = cesavedir + "/" + savename
    plt.savefig(savedir)
    
for i in tqdm(range(len(tempas))):
    ns = np.load(tempas[i])
    S_DB = librosa.power_to_db(ns, ref=np.max)
    librosa.display.specshow(S_DB,sr=10000, hop_length=512, 
                             x_axis='time', y_axis='mel');
    savename = tempas[i].split("/")[-1].split(".")[0] + ".png"
    savedir = assavedir + "/" + savename
    plt.savefig(savedir)
```
