```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

```

```python
dir1 = 'C:/Users/MAKinteract/makinteract/ur/ur_33_32.npy' # nothing
ns = np.load(dir1)
```

```python
print(ns.shape)
S_DB = librosa.power_to_db(ns, ref=np.max)
librosa.display.specshow(S_DB,sr=10000, hop_length=512, 
                         x_axis='time', y_axis='mel');
plt.colorbar(format='%+2.0f dB');
```
